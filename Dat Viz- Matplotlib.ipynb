```python
x= np.linspace(-np.pi,np.pi,1000)
plt.plot(x, 1/(1+np.exp(-x))) # Sigmoide / função logística
```

```python
plt.plot(x,np.abs(x)) #módulo
```

```python
plt.figure(figsize=(5,5)) # Define parâmetros visuais do gráfico, figsieze diz o tamanho
plt.plot(x,np.abs(np.sin(x))) # Módulo do sen de x
plt.plot(x,np.sin(x)) # função seno
plt.show() # Mostra tudo o que tiver entre ele e o plt. figure. basicamente o gráfico
# Aqui plotamos as duas funções em uma só
```

```python
# Para salvar gráfico
figura = plt.figure(figsize=(3,4)) # Primeiro atribui o figure a um objeto/variável
plt.plot(x,x**2)

figura.savefig('figura1.png') # Método que salva passando o caminho de onde salvar, mas existem outros argumentos de estilização
```

```python
plt.style.use('seaborn-colorblind') # Aplica um estilo pré pronto no gráfico. nesse caso um filtro para anular daltonismo
x = np.linspace(-2*np.pi,2*np.pi,1000)
figura2 =  plt.figure(figsize=(10,4))
plt.title('Função seno e cosseno',fontsize = 16) # adiciona título e fontsize diz o tamanho

plt.plot(x,np.sin(x), label ='f(x) = sen(x)',ls = '--') # Label as legendas para aquela função / curva. mas precisa ser ativado pelo plt.legend()
plt.plot(x,np.cos(x), label = 'f(x) = cos(x)', ls =':', lw = 4, color = '#be23cc') # o argumento ls diz se é pontilhado, tracejado... 
# lw = largura da linha
#color diz a  a cor


plt.xlabel('X')
plt.ylabel('F(x)')

plt.legend(bbox_to_anchor = (1,0.5), loc ='center left') # Ativa a legenda no gráfico # os argumentos mudam posição da legenda no plot

plt.show()
figura2.savefig('Seno e cosseno.png', bbox_inches = 'tight') # Bbox_inches altera a espessura do traço da borda
```

```python
# Plotando vários gráficos separados
plt.style.use('default') # Como foi usado anteriormente precisamos voltar os estilo para default

plt.subplot(221) # Divide os gráficos, recebe 3 argumentos em um só numero: i e j da matriz indicando o tamnho total e a posição do gráfico a seguir
plt.plot(x,np.sin(x), color ='#47108f') 
plt.title('Seno')
plt.xlabel('x')
plt.ylabel('f(x) = sen(x)')

plt.subplot(222)
plt.plot(x,np.cos(x))
plt.title('Cosseno')
plt.xlabel('x')
plt.ylabel('f(x) = cos(x)')


plt.subplot(223)
plt.plot(x,np.tanh(x)) 
plt.title('Tangente hiperbólica')
plt.xlabel('x')
plt.ylabel('f(x) = tanh(x)')


plt.subplot(224)
plt.plot(x,np.exp(x), color ='#47108f')
plt.title('Exponencial')
plt.xlabel('x')
plt.ylabel('f(x) = exp(x)')



plt.tight_layout() # Ajusta o layout para que tudo se encaixe proporcionalmente
```
